```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('Copy of sonar data.csv',header=None)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.replace('R',1,inplace=True)
df.replace('M',0,inplace=True)
```

```python
df.info
```

```python
df.shape
```

```python
df.describe()
```

```python
df.groupby(60).mean()
```

```python
x=df.drop([60],axis=1)
```

```python
y=df[60]
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y, random_state=5)
```

```python
xgb=XGBClassifier()
```

```python
y_train
```

```python
xgb.fit(x_train,y_train)
```

```python
y_pred=xgb.predict(x_test)
```

```python
score1=accuracy_score(y_pred,y_test)
```

```python
score1
```

```python
# Random forest classifier
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred1=rf.predict(x_test)
```

```python
score2=accuracy_score(y_pred1,y_test)
```

```python
score2
```

```python
# KNN
knn= KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred2=knn.predict(x_test)
```

```python
score3=accuracy_score(y_pred2,y_test)
```

```python
score3
```

```python
bag=BaggingClassifier()
```

```python
bag.fit(x_train,y_train)
```

```python
y_pred3=bag.predict(x_test)
```

```python
score4=accuracy_score(y_pred3,y_test)
```

```python
score4
```

```python
param_grid = {
    'n_estimators': [50,100,500], 
    'max_samples': [0.1,0.4,0.7,1.0],
    'bootstrap' : [True,False],
    'max_features' : [0.1,0.4,0.7,1.0]
    }
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
search=GridSearchCV(estimator=BaggingClassifier(),param_grid=param_grid,n_jobs=-1,cv=5)
```

```python
search.fit(x_train,y_train)
```

```python
search.best_score_
```

```python
search.best_params_ 
```

```python
# now you can see that we can get accuracy score of 0.85 by beggiong classifier
```

```python
bag1=BaggingClassifier(bootstrap= False,
                       max_features= 0.1,
                       max_samples=0.7,
                        n_estimators= 500)
```

```python
bag1.fit(x_train,y_train)
```

```python
y_pred4=bag1.predict(x_test)
```

```python
score5=accuracy_score(y_pred4,y_test)
```

```python
score5
```

```python
# now you can see our accuracy score is improved 
```

```python
lr =LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred5=lr.predict(x_test)
```

```python
accuracy_score(y_pred5,y_test)
```
